```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import folium

url = 'https://www.geographyrealm.com/map-and-geography-trivia-u-s-national-parks/'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

tbl = soup.find("table")

data_frame = pd.read_html(str(tbl))[0]
```

```python
data_frame
```

```python
data_frame.columns = data_frame.iloc[0]
data_frame = data_frame[1:]
```

```python
data_frame
```

```python
data_frame.info()
```

```python
data_frame['Latitude'] = pd.to_numeric(data_frame['Latitude'])
data_frame['Longitude'] = pd.to_numeric(data_frame['Longitude'])
data_frame.info()
```

```python
data_frame.dropna(subset=['Latitude', 'Longitude'])
```

```python
data_frame.info()
```

```python
mean_latitude = data_frame.Latitude.mean()
mean_longitude = data_frame.Longitude.mean()

data_frame['Latitude'].fillna(mean_latitude, inplace=True)
data_frame['Longitude'].fillna(mean_longitude, inplace=True)
```

```python
mymap_2 = folium.Map(location=[data_frame.Latitude.mean(), data_frame.Longitude.mean()], zoom_start=12)
for row in data_frame[['Latitude','Longitude', 'Park', 'Park Established']].values:
    folium.CircleMarker(location=[row[0], row[1]], radius=1,color='red', popup=["Park Name", row[2], "Park Established", row[3]]).add_to(mymap_2)
```

```python
folium.TileLayer('Stamen Water Color').add_to(mymap_2)
folium.TileLayer('Stamen Terrain').add_to(mymap_2)
folium.LayerControl().add_to(mymap_2)
```

```python
mymap_2.save('established_dates.html')
```
